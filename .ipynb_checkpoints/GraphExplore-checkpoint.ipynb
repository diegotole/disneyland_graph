```python
short_stop = 100
solutions = []
hops = 0

class DFS:
    
    mypath = []
    mypath_set = set()
    solutions = []
    short_stop = 80
    hops = 0
    stop_search = 5_000_000

    @classmethod
    def add_solution(cls):
        cls.solutions(
            list( cls.mypath )
        )
    
    @classmethod    
    def should_stop(cls):
        if len(cls.mypath) > cls.short_stop:
            return True
        
        return False
    
    @classmethod
    def pop_path(cls):
        if len(cls.mypath):
            k = cls.mypath.pop()
            cls.mypath_set.remove(k)
            
    @classmethod
    def append_path(cls, k):
        cls.mypath.append(k)
        cls.mypath_set.add(k)
    
    @classmethod
    def jump(cls, mymap, mypath,  curr_key):
        
        cls.hops+=1
        if cls.hops >= cls.stop_search:
            #raise Exception("TOO LONG HERE")
            return
        
        if cls.should_stop():
            cls.pop_path()
            return

        if cls.hops % 1_000_000 == 0:
            print( cls.hops ,curr_key , len(mypath), len(solutions) )

        if len(set(mypath)) == TOTAL_ROWS:
            cls.add_solution()
            
            cls.pop_path()
            return



        for k in mymap[curr_key]:

            if k in cls.mypath_set:
                continue

            cls.append_path(k)
            cls.jump(mymap, mypath, k)

        cls.pop_path()

        return
        
        
```
